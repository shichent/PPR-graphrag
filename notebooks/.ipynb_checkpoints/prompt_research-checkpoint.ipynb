````python
import re

def make_format_and_fill(template: str, values: list[str]) -> str:
    # Replace each [...] with the corresponding value
    i = 0
    def replacer(m):
        nonlocal i
        result = values[i]
        i += 1
        return result
    
    return re.sub(r'\[.*?\]', replacer, template)
````

````python
import re

def find_tags(tag_type:str,text: str) -> list[str]:
    """
    Finds all occurrences of strict patterns like [PERSON#X] 
    where X is a number between 1 and 9999.
    """
    pattern = fr'{tag_type}#[1-9]\d{{0,3}}'
    
    # re.findall returns a list of all non-overlapping matches in the string
    return re.findall(pattern, text)
# entities = {}
# for x in find_tags('PERSON',documents[i]['text']):
#     entities[x] = 'person'
# for x in find_tags('LOCATION',documents[i]['text']):
#     entities[x] = 'location'
# entities
````

````python
import networkx as nx
import asyncio
import json_repair
G = nx.MultiGraph()
llm = LLMCompletionCall('deepseek-chat')
isolated = set()
# Load triples into the graph
# chunks = [text1[1000*i:1000*i+1200] for i in range(20)]

chunks = documents
prompts = []
llm = LLMCompletionCall()
for i in range(200,220):
    entities = {}
    for x in find_tags('PERSON',documents[i]['chunk']):
        entities[x] = 'person'
    for x in find_tags('LOCATION',documents[i]['chunk']):
        entities[x] = 'location'
    prompt = prompt_fix_anony.format(chunk=chunks[i],schema = schema_llm)
    prompts.append(prompt)
tasks = [llm.async_call_api(prompt) for prompt in prompts]
results = await asyncio.gather(*tasks)
for response in results:
    d = json_repair.loads(response)
    for subject, predicate, obj in d['triples']:
        G.add_edge(subject, obj, relation=predicate)
    for entity in d['attributes'].keys():
        if entity not in G.nodes():
            isolated.add(entity)

print(f"Number of components: {nx.number_connected_components(G)}. Size of each component:")
for component in nx.connected_components(G):
    print(len(component))
print(f"Number of isolated attributes: {len(isolated)}")
print(f"Total number of nodes:{len(G.nodes())}, edges: {len(G.edges())}, ration: {len(G.edges())/len(G.nodes())}")

````

````python
from gliner import GLiNER

# Initialize GLiNER with the base model
model = GLiNER.from_pretrained("urchade/gliner_medium-v2.1")
````

````python
import re

# Sample text for entity prediction
text = """
请查找关于《水浒传》中人物[PERSON#924]的宝刀被没收的具体情节，包括没收的机构是什么。
"""

##### Unicode separating
text = re.sub(r'([\u0080-\uFFFF])', r' \1', text)
````

````python


# Labels for entity prediction
# Most GLiNER models should work best when entity types are in lower case or title case
labels = [    "person",
    "location",
    "organization",
    "event",
    "object",
    "concept",
    "time"]

# Perform entity prediction
entities = model.predict_entities(text, labels, threshold=0.05)

ans = []
# Display predicted entities and their labels
for entity in entities:
    ans.append(re.sub(r'(?<=[\u0080-\uFFFF]) (?=[\u0080-\uFFFF])', '', entity["text"])+" => "+entity["label"])
````

````python
def transform_finer_data(examples):
    target_output = []
    for text, entities in zip(examples["text"], examples["char_spans"]):
        # 1. Apply your modifications to the targets
        processed_labels = []
        for ent in entities:
            s,e,l = ent['start'],ent['end'],ent['label']
            processed_labels.append({'entity_name':text[s:e],'label':l})
        
        target_output.append(str(processed_labels))

    return {"text": examples["text"],'target':target_output}
````

````python
total_true = 0
total_exp = 0
total_correct_label = 0
total_match = 0
results = []
for i in range(20,30):
    d = dataset_chinese['train'][i]
    provided_text = d['text']
    true_label = {}
    for dd in d['char_spans']:
        s,e,l = dd['start'],dd['end'],dd['label']
        true_label[provided_text[s:e]] = l
    response = llm.call_api(prompt_span.format(provided_text = provided_text, example_response = example_response))
    r_json = json.loads(response)
    if isinstance(r_json,dict):
        r_dict = r_json
    else:
        r_dict = r_json[0]
    total_true += len(true_label)
    total_exp += len(r_dict)
    match = 0
    correct_label = 0
    for k in r_dict:
        if k in true_label:
            match += 1
            if true_label[k] == r_dict[k]:
                correct_label += 1
    total_match += match
    total_correct_label += correct_label
    print(f"Model found {match} of {len(true_label)} entities. {correct_label} of them are labeled correctly.")
    results.append({})
    results[-1]['predict'] = r_dict
    results[-1]['truth'] = true_label
print(f"Total extracted count = {total_exp}. Total true count = {total_true}")
print(f"Correct extract rate = {total_match/total_true}")
print(f"Correct label rate = {total_correct_label/total_true}")
````

````python
import json

def load_red_dataset(data_path, rel_info_path=None):
    """
    Parses DocRED/Re-DocRED/HacRED format into LLM-friendly text chunks and triples,
    including 'alias of' relations for co-references.
    """
    HacRED = False
    if 'HacRED' in data_path:
        HacRED = True
    # 1. Load Relation Mapping
    rel_map = {}
    if rel_info_path:
        with open(rel_info_path, 'r', encoding='utf-8') as f:
            rel_map = json.load(f)
            
    # 2. Load the main dataset
    if HacRED:
        raw_data = []
        with open(data_path, 'r', encoding='utf-8') as f:
            for line in f:
                raw_data.append(json.loads(line))
    else:
        with open(data_path, 'r', encoding='utf-8') as f:
            raw_data = json.load(f)
        
    llm_dataset = []
    
    # 3. Process each document chunk
    for doc in raw_data:
        # A. Reconstruct the Text Chunk
        sentences = [" ".join(sent) for sent in doc.get("sents", [])]
        full_text = " ".join(sentences)
        if HacRED:
            doc['vertexSet'] = doc['vertex_char']
            full_text = doc['text']
            doc['labels'] = doc['labels_char']
        
        entities_dict = {}
        entities_byid = {}
        all_triples = []
        for i in range(len(doc['vertexSet'])):
            v_list = doc['vertexSet'][i]
            for v in v_list:
                idx = i
                entities_dict[v['name']] = v['type']
                if idx not in entities_byid:
                    entities_byid[idx] = [v['name']]
                else:
                    if v['name'] not in entities_byid[idx]:
                        entities_byid[idx].append(v['name'])

        for idx in entities_byid:
            for i in range(1,len(entities_byid[idx])):
                entities_list = entities_byid[idx]
                all_triples.append((entities_list[i],'alias of',entities_list[0]))

        for e in doc['labels']:
            head = entities_byid[e['h']][0]
            tail = entities_byid[e['t']][0]
            if HacRED:
                rel = e['r']
            else:
                rel = rel_map[e['r']]
            all_triples.append([head,rel,tail])
        
        formatted_row = {
            "text": full_text,
            "entity_types": str(entities_dict),
            "triples": all_triples
        }
        
        llm_dataset.append(formatted_row)
        
    return llm_dataset
````

````python
def pack_lines(input_file, output_file):
    with open(input_file, 'r') as f_in, open(output_file, 'w') as f_out:
        buffer = []
        
        for line in f_in:
            # Check if the line starts with the delimiter
            if line.startswith('id: ') and buffer:
                # Join the previous block and write it out
                f_out.write(" ".join(buffer) + "\n")
                buffer = []
            
            # Strip whitespace to avoid double spaces/newlines
            clean_line = line.strip()
            if clean_line:
                buffer.append(clean_line)
        
        # Don't forget to write the very last block in the file
        if buffer:
            f_out.write(" ".join(buffer) + "\n")

# Usage
pack_lines(r'output\chunks\2.txt', r'output\chunks\2_cleaned.txt')
````
